```python
from ete3 import NCBITaxa
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

ncbi = NCBITaxa()
```

```python
kaiju_out = pd.read_csv('kaiju_out/nirk_meta_good_greedy',sep='\t',header=0,names=['classified_unclassified','read_name','taxid','score','taxid2','seqid','seq'])
# kaiju_out = pd.read_csv('kaiju_out/nirk_meta_all_greedy',sep='\t',header=0,names=['classified_unclassified','read_name','taxid','score','taxid2','seqid','seq'])
```

Retrieve pHs for each read.

```python
pH = np.zeros(len(kaiju_out))
for i in range(0,len(kaiju_out)):
    start = kaiju_out['read_name'][i].find('pH_')
    end = kaiju_out['read_name'][i].find('_NODE')
    pH[i] = float(kaiju_out['read_name'][i][start+3:end])
```

Retrieve class of each best match

```python
tax_level = 4
cla = np.zeros(len(kaiju_out))
for i in range(0,len(kaiju_out)):
    if kaiju_out['classified_unclassified'][i] == 'C': #if unclassified, zero
        lin = ncbi.get_lineage(kaiju_out['taxid'][i])
        if len(lin)>tax_level+1: #if poorly unclassified, zero
            cla[i] = lin[tax_level]
cla_id = ncbi.get_taxid_translator(cla) #retrieve names of phylogenetic classes
```

Count taxa in each category

```python
classes = list(set(cla))
cla_ct = np.zeros(len(classes))
cla_lab = []

cla_ct[0] = np.sum(cla==classes[0])
cla_lab.append('Unclassified')
for i in range(1,len(classes)):
    cla_ct[i] = np.sum(cla==classes[i])
    cla_lab.append(cla_id[classes[i]])
    
plt.figure(figsize=(6, 4), dpi=80)
plt.bar(np.arange(0,len(classes)),cla_ct);
plt.xticks(np.arange(0,len(classes)),cla_lab,rotation=45,horizontalalignment='right',verticalalignment='center_baseline');
plt.ylabel('counts');
```

Break out relative abundances as a function of pH

```python
cla_hiabund = np.where(cla_ct>75)[0]; #focus on the high-abundance classes

nbins = 3
edges = np.append(np.append(np.min(pH), np.quantile(pH,np.arange(1,nbins)/nbins)),np.max(pH))
bin_lab = []

cla_ct = np.zeros((nbins,len(classes)))
total_ct = np.zeros(nbins)
for i in range(0,nbins):
    idx = np.where(np.logical_and(pH>=edges[i],pH<edges[i+1]))[0]
    bin_lab.append(str(edges[i]) + '-' + str(edges[i+1]))
    cla_sub = cla[idx]
    total_ct[i] = len(cla_sub)
    for j in range(0,len(classes)):
        cla_ct[i,j] = np.sum(cla_sub==classes[j])
```

Focus on the high abundance taxa

```python
other = total_ct-np.sum(cla_ct[:,cla_hiabund],axis=1)
other[...,None]
cla_hiabund_ct = np.append(cla_ct[:,cla_hiabund],other[...,None],axis=1)

cla_hiabund_lab = []
for i in range(0,len(cla_hiabund)):
    cla_hiabund_lab.append(cla_lab[cla_hiabund[i]])
cla_hiabund_lab.append('Other')

for i in range(0,len(cla_hiabund)+1):
    plt.bar(np.arange(0,nbins),cla_hiabund_ct[:,i]/total_ct,bottom=np.sum(cla_hiabund_ct[:,0:i],axis=1)/total_ct)
plt.legend(cla_hiabund_lab,bbox_to_anchor = (1.05, 0.6))
plt.xticks(np.arange(0,nbins),bin_lab);
plt.xlabel('soil pH (1 M KCl method)');
plt.ylabel('% gene content');
```
