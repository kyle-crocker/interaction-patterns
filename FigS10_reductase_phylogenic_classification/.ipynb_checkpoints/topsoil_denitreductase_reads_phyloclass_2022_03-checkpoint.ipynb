```python
from ete3 import NCBITaxa
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pickle5 as pickle
import seaborn as sns
import matplotlib.patches as mpatches

ncbi = NCBITaxa()
```

Read data

```python
#output of Kaiju server for taxonomic identification of reads
kaiju = pd.read_csv('kaiju_out/denitKOs_meta_all_greedy.out',sep='\t',header=0,names=['classified_unclassified','read_name','taxid','score','taxid2','seqid','seq'])

#Lookup table associating SRA IDs with sample names
run_labels = pd.read_csv('sample_run_labels.csv')

#Table of environmental metadata
env_vars = pd.read_csv('env_vars.csv')

#Table relating denitrification KOs to gene names
denit_KOs = pd.read_csv('denit_kos.tsv',sep='\t',header=None,names=['KO','gene'])

#Reads from topsoil microbiome dataset that map to ORFs annotated as denitrification genes
with open('20220214_to_karna_reads.pk', 'rb') as f:
    reads = pickle.load(f)
    
#ORFs assembled for topsoil microbiome dataset that are annotated as denitrification genes
with open('20220214_to_karna_orfs.pk', 'rb') as f:
    orfs = pickle.load(f)
```

Match each read with a pH value, denit gene label, and taxonomic id

```python
ph = []
sample = []
run = []
gene = []
all_tax = []        
get_ranks = ['superkingdom','phylum','class','order','family','genus']

for i in range(0,len(kaiju)):
    read_name = kaiju['read_name'][i]
    
    #match each read with a pH value
    run_name = read_name[:read_name.find('.')]
    sample_name = run_labels['sample'][run_labels['run'] == run_name].values[0]
    ph.append(env_vars['pH'][env_vars['sample']==sample_name].values[0])
    run.append(run_name)
    sample.append(sample_name)
    
    #match each read with a denit gene name
    orf_id = reads['orf_id'][reads['read_id'] == read_name].values[0]
    KO = orfs['KO'][orfs['orf_id']==orf_id].values[0].split(",")
    for j in range(0,len(KO)):
        KO[j] = KO[j][KO[j].find(":")+1:]
        gene_name = denit_KOs['gene'][denit_KOs['KO'] == KO[j]]
        if gene_name.size > 0:
            gene.append(gene_name.values[0])
    
    #look up taxonomic information for each classified read
    if kaiju['classified_unclassified'][i] == 'C':
        lin = ncbi.get_lineage(kaiju['taxid'][i])
        ranks = ncbi.get_rank(lin);
        inv_ranks = {}
        for k, v in ranks.items():
            inv_ranks[v] = inv_ranks.get(v, []) + [k]
            
        tax_str = []
        for j in range(0,len(get_ranks)):
            try:
                rank_idx = lin.index(inv_ranks[get_ranks[j]][0])
                tax = ncbi.get_taxid_translator([lin[rank_idx]])
                tax_str.append(tax[list(tax)[0]])
            except:
                tax_str.append('')
    else:
        tax_str = ['','','','','','']
    all_tax.append(tax_str)
        
kaiju['ph'] = ph
kaiju['gene'] = gene
kaiju['sample'] = sample
kaiju['run'] = run

all_tax = np.array(all_tax)
for i in range(0,len(get_ranks)):
    kaiju[get_ranks[i]] = all_tax[:,i]
```

Plot count of genes in dataset. Why are there so few NarG genes? In original dataset, NarG was 27% of genes. Here it's 16%

```python
genes = np.unique(kaiju['gene'])
genes_ct = []
for i in range(0,len(genes)):
    genes_ct.append(np.sum(kaiju['gene']==genes[i]))
    
plt.bar(genes, genes_ct)
plt.ylabel('count');

genes_ct[1]/np.sum(genes_ct)
```

Plot something similar to what we obtained before by looking at the KO counts from the global topsoil microbiome dataset. Seems like the trend is still there but perhaps noisier.

```python
runs = np.unique(kaiju['run'].values)
genes = np.unique(kaiju['gene'])

#get phs for runs
phs = []
for run in runs:
#     temp_phs = kaiju['ph'].loc[kaiju['run']==run].values[0]
    phs.append(kaiju['ph'].loc[kaiju['run']==run].values[0])
phs = np.array(phs)

#count reductase genes in for each run
run_gene_ct = pd.DataFrame(0,index = runs, columns=genes)
for run in runs:
    run_idx = kaiju['run'] == run
    for gene in genes:
        gene_idx = kaiju['gene'] == gene
        run_gene_ct.loc[run,gene] = len(np.where(np.logical_and(run_idx,gene_idx))[0])

#drop nor for sake of comparison with previous result
run_gene_ct = run_gene_ct.drop(columns=['nor'])
run_gene_ct = run_gene_ct.reindex(columns=['narG','napA','nirS','nirK','nosZ'])

#calculate percentages
run_gene_pct = run_gene_ct.copy()
run_gene_pct = run_gene_pct.divide(np.sum(run_gene_ct,axis=1),axis=0)
G = run_gene_pct.values

#remove observations with too few gene counts for precise estimation of fractions
runs_keep = np.sum(run_gene_ct.values,axis=1) > 80
G = G[runs_keep,:]
phs = phs[runs_keep]

#sort rows by pH (increasing)
G = G[np.argsort(phs),:]

#plot
width = 1
p1 = plt.bar(range(0,len(G)), G[:,4], width, color='g')
p2 = plt.bar(range(0,len(G)), G[:,3], width, bottom=G[:,4], color='r')
p3 = plt.bar(range(0,len(G)), G[:,2], width, bottom=G[:,4]+G[:,3], color='y')
p4 = plt.bar(range(0,len(G)), G[:,1], width, bottom=G[:,4]+G[:,3]+G[:,2], color='c')
p5 = plt.bar(range(0,len(G)), G[:,0], width, bottom=G[:,4]+G[:,3]+G[:,2]+G[:,1], color='b')
plt.xlabel('sample')
plt.ylabel('relative abundance')

patch1 = mpatches.Patch(color='b', label='narG')
patch2 = mpatches.Patch(color='c', label='napA')
patch3 = mpatches.Patch(color='y', label='nirS')
patch4 = mpatches.Patch(color='r', label='nirK')
patch5 = mpatches.Patch(color='g', label='nosZ')
plt.legend(handles=[patch1,patch2,patch3,patch4,patch5],bbox_to_anchor = (1.05, 1.0))
plt.show()

plt.scatter(np.sort(phs),G[:,0])
plt.xlabel('pH')
plt.ylabel('narG relative abundance')
plt.show()
```

For each gene, look at breakdown of taxa in different pH bins. Problem with this approach is that samples with large numbers of reads contribute heavily.

```python
genes = np.unique(kaiju['gene'])

for gene in genes:
    org = 'class'

    gene_idx = kaiju['gene']==gene
    categories = list(np.unique(kaiju[org][gene_idx]))

    nbins = 3
    edges = np.append(np.append(np.min(kaiju['ph'].values), np.quantile(kaiju['ph'].values,np.arange(1,nbins)/nbins)),np.max(kaiju['ph'].values))
    bin_lab = []
    #create text labels for each bin
    for i in range(0,nbins):
        bin_lab.append(str(edges[i]) + '-' + str(edges[i+1]))

    #record counts in each category
    cat_ct = pd.DataFrame(0,index=categories,columns=bin_lab) #dataframe to store counts
    for i in range(1,len(edges)):
        if i == len(edges):
            ph_idx = np.logical_and(kaiju['ph']>=edges[i-1],kaiju['ph']<=edges[i])
        else:
            ph_idx = np.logical_and(kaiju['ph']>=edges[i-1],kaiju['ph']<edges[i])
        idx = np.where(np.logical_and(ph_idx,gene_idx))[0]
        for j in range(0,len(idx)):
            cat_ct.loc[kaiju[org][idx[j]],bin_lab[i-1]] += 1 

    cat_ct = cat_ct.rename(index={'': 'Unclassified'});

    #create a percentages dataframe
    cat_pct = cat_ct.copy()
    cat_pct = cat_pct.divide(np.sum(cat_ct),axis=1)

    #create a summarized percentages dataframe
    min_pct = 0.05;
    rm_row = []
    other = pd.DataFrame(0,index=['Other'],columns = bin_lab)
    for i in range(0,len(cat_pct)):
        if np.max(cat_pct.iloc[i]) < min_pct:
            rm_row.append(cat_pct.iloc[i].name)
            other += cat_pct.iloc[i]
    cat_pct_summ = cat_pct.copy()
    cat_pct_summ = cat_pct_summ.drop(index=rm_row)
    cat_pct_summ = cat_pct_summ.append([other])

    for i in range(0,len(cat_pct_summ)):
        plt.bar(np.arange(0,nbins),cat_pct_summ.iloc[i],bottom=np.sum(cat_pct_summ.iloc[0:i]))
    plt.legend(cat_pct_summ.index,bbox_to_anchor = (1.05, 1.0))
    plt.xticks(np.arange(0,nbins),bin_lab);
    plt.xlabel('soil pH (1 M KCl method)');
    plt.ylabel('% gene content');
    plt.title(gene);
    plt.show()
```
